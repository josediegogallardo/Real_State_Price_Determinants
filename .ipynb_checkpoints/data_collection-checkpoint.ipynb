```python
import datetime
import requests

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import statsmodels.api as sm
#import pandas_datareader as pdr


from math import sqrt
from statsmodels.tsa.filters.hp_filter import hpfilter
```

### <a id='dif'> Quarterly data from BCRP </a>

```python
# Basic parameters
base_url   = "https://estadisticas.bcrp.gob.pe/estadisticas/series/api/"
formato    = "/json"
period     = "/2008-1/2023-4"
quarter_id = ["PD37940PQ", #Hedonic Real Estate Price Index
              "PN02538AQ"] #Gross Domestic Product - GDP
id_name    = ["Real State Price Index",
              "Gross Domestric Product"]

# Series as valid format for request
quarter_series = '-'.join(quarter_id)

# Information Request
quart_data_json = requests.get(base_url + quarter_series + formato + period).json()
quart_data = pd.DataFrame(quart_data_json["periods"])

# Values data
quart_data[id_name] = pd.DataFrame(
                          quart_data["values"].tolist(),
                          index = quart_data.index).astype(float)
quart_data = quart_data.drop('values', axis=1)

## Dates data
# Source date format
quart_data = quart_data.rename(columns = {"name": "origin_dates"})
# Python date format
date = pd.date_range('1/1/2008',
                     periods = quart_data.shape[0],
                     freq ='q')
quart_data['date'] = pd.to_datetime(date, errors='coerce')
quart_data.set_index('date', inplace=True)
quart_data.tail()
```

```python
quart_data.plot(subplots = True,
                figsize = (15,5),
                layout = (1,2))
```
